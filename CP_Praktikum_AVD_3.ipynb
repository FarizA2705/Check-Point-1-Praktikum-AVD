### **Outliers Values**

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

## **Outliers**

Outlier merupakan nilai yang berbeda dengan nilai yang lain dalam dataset, sebuah nilai outliers bisa jadi lebih rendah atau lebih tinggi. Outlier terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lainnya.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

1. **monthly_visits**




```python
Q1 = df['monthly_visits'].quantile(0.25)
Q3 = df['monthly_visits'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['monthly_visits'] = df['monthly_visits'].clip(lower=lower_bound, upper=upper_bound)
```

2. **avg_session_duration_s**

```python
Q1 = df['avg_session_duration_s'].quantile(0.25)
Q3 = df['avg_session_duration_s'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['avg_session_duration_s'] = df['avg_session_duration_s'].clip(lower=lower_bound, upper=upper_bound)
```

3. **stickiness_index**

```python
Q1 = df['stickiness_index'].quantile(0.25)
Q3 = df['stickiness_index'].quantile(0.75)

IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df['stickiness_index'] = df['stickiness_index'].clip(lower=lower_bound, upper=upper_bound)
```

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Outlier atau data pencilan pada dataset ini dibersihkan menggunakan metode IQR (Interquartile Range) untuk menjaga objektivitas analisis. Proses ini dilakukan dengan menghitung rentang antara kuartil atas (Q3) dan kuartil bawah (Q1) untuk menentukan "pagar" batas kewajaran trafik web. Semua domain yang memiliki kunjungan bulanan di atas batas tersebut, seperti raksasa teknologi https://www.google.com/search?q=google.com atau facebook.com, akan dihapus dari dataset. Pembersihan ini bertujuan agar data menjadi lebih seragam dan tidak timpang, sehingga nilai rata-rata yang dihasilkan mencerminkan performa situs web pada umumnya tanpa terganggu oleh angka-angka ekstrem dari domain yang terlalu mendominasi pasar.

# **Construct Data**

Tahap Construct Data adalah proses kreatif dalam analisis data. Di sini, kita tidak lagi hanya membersihkan data, melainkan membuat sesuatu yang baru dari data yang sudah ada untuk memperkaya analisis.

1. **Membuat kolom baru**

Kolom yang ditambahkan adalah "Effective Engaging Visits." Saya menambahkan kolom tersebut karena variabel tersebut sangat penting untuk metrik kunjungan total saja sering menipu; sebuah situs bisa memiliki jutaan klik namun pengunjungnya langsung pergi tanpa melakukan interaksi apapun. dengan ditambahkannya variabel tersebut, kita dapat mengukur kualitas sebuah trafik secara nyata dan memberikan penilaian performa yang adil bagi situs-situs yang memiliki trafik yang lebih kecil namun memiliki tingkat keterlibatan pengguna yang tinggi.

```python
def categorize_engagement(index):
    if index > 15:
        return 'High Engagement'
    elif index >= 10:
        return 'Standard Engagement'
    else:
        return 'Low Engagement'

df['Engagement Status'] = df['stickiness_index'].apply(categorize_engagement)
```
